```python
# Find titles with more than one occurrence
duplicate_titles = df_sorted['title'].value_counts()
duplicate_titles = duplicate_titles[duplicate_titles > 1]

print(duplicate_titles)
print("="*1000)
# Print all rows with duplicate titles
duplicate_rows = df_sorted[df_sorted['title'].isin(duplicate_titles.index)]

print(duplicate_rows)
```

```python
duplicate_rows
```

```python
# Step 1: Drop duplicate descriptions (keep first occurrence)
df_sorted = df_sorted.drop_duplicates(subset=['description'], keep='first')


# Step 2: Merge rows having the same title
def merge_descriptions(group):
    # Combine all descriptions into one string
    merged_description = " ".join(group['description'].dropna().astype(str))

    # Keep the first row's details
    row = group.iloc[0].copy()
    row['description'] = merged_description

    return row


# Apply merging based on title
df_sorted = (
    df_sorted
    .groupby('title', as_index=False, group_keys=False)
    .apply(merge_descriptions)
)


# Reset index after merging
df_sorted = df_sorted.reset_index(drop=True)
```

```python
# Find titles with more than one occurrence
duplicate_titles = df_sorted['title'].value_counts()
duplicate_titles = duplicate_titles[duplicate_titles > 1]

print(duplicate_titles)
print("="*1000)
# Print all rows with duplicate titles
duplicate_rows = df_sorted[df_sorted['title'].isin(duplicate_titles.index)]

print(duplicate_rows)
```
