## Automated Pipeline for fetching imbd data selected movies from jiohotstar

```python
# Install tqdm for a beautiful progress bar in Google Colab
!pip install -q tqdm

import pandas as pd
import requests
import time
import json
import os
from tqdm.notebook import tqdm
```

```python
# --- CONFIGURATION ---
API_KEY = "7146fa93"
OMDB_URL = "https://www.omdbapi.com/"
EXCLUDE_KEYS = {'Title', 'Year', 'Response'} # Fields to exclude from API response
DELAY_BETWEEN_REQUESTS = 0.3 # Seconds to prevent hitting OMDb rate limits (1000 req/day)


df = pd.read_csv('/content/Jiohostar_preprocessedData.csv')
print(f"Original DataFrame shape: {df.shape}")
df.head()
```

```python
def fetch_movie_data(row):
    # Safely extract values, handling NaNs
    movie_name = str(row['title']).strip() if pd.notna(row['title']) else ""
    year = str(row['released_year']).strip() if pd.notna(row['released_year']) else None

    if not movie_name:
        return {'api_error': 'Missing movie title'}

    params = {
        "apikey": API_KEY,
        "t": movie_name,
        "plot": "full",
        "r": "json"
    }

    # Optional: Add year to improve search accuracy
    if year and year.lower() != 'nan':
        params["y"] = year

    try:
        # Add timeout to prevent hanging
        response = requests.get(OMDB_URL, params=params, timeout=10)
        response.raise_for_status() # Raise exception for HTTP errors (4xx, 5xx)

        movie_data = response.json()

        if movie_data.get('Response') == 'True':
            # Filter out the excluded keys
            filtered_data = {k: v for k, v in movie_data.items() if k not in EXCLUDE_KEYS}

            # Flatten complex types (lists/dicts like 'Ratings') to JSON strings for safe CSV saving
            for key, value in filtered_data.items():
                if isinstance(value, (list, dict)):
                    filtered_data[key] = json.dumps(value)

            return filtered_data

        else:
            # API responded, but movie was not found
            return {'api_error': movie_data.get('Error', 'Movie not found')}

    except requests.exceptions.Timeout:
        return {'api_error': 'Request timed out'}
    except requests.exceptions.RequestException as e:
        return {'api_error': f"Network error: {str(e)}"}
    except Exception as e:
        return {'api_error': f"Unexpected error: {str(e)}"}
```

```python
# Enable tqdm for pandas apply
tqdm.pandas(desc="Enriching Data")

print("Starting API enrichment pipeline...")
# Apply the function to each row
api_results = df.progress_apply(fetch_movie_data, axis=1)

# Introduce a small delay batch-wise if you have thousands of rows
# (Optional but recommended for large datasets to respect rate limits)
time.sleep(DELAY_BETWEEN_REQUESTS)
```

```python
# Convert the list of dictionaries into a new DataFrame
results_df = pd.DataFrame(api_results.tolist(), index=df.index)

# Concatenate original data with new API data side-by-side
enriched_df = pd.concat([df, results_df], axis=1)

print(f"\nEnriched DataFrame shape: {enriched_df.shape}")
print("Columns added:", [col for col in enriched_df.columns if col not in df.columns])

# Display a sample to verify
enriched_df.head()
```

## Saving in those poster in google drive

```python
import os
import requests
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
from urllib.parse import urlparse
import time

# ============================================
# STEP 1: Setup Directory Structure
# ============================================
poster_folder = "/content/drive/MyDrive/poster"

# Create directory if it doesn't exist
if not os.path.exists(poster_folder):
    os.makedirs(poster_folder)
    print(f"✅ Created directory: {poster_folder}")
else:
    print(f"📁 Directory already exists: {poster_folder}")

# ============================================
# STEP 2: Generate Unique 4-Digit Codes
# ============================================
# Create a mapping of titles to unique 4-digit codes
enriched_df['imageid'] = None
enriched_df['format'] = None

# Assign unique 4-digit codes to each row
for idx in range(len(enriched_df)):
    # Generate 4-digit code with zero padding (0001, 0002, etc.)
    code = f"{idx + 1:04d}"
    enriched_df.loc[idx, 'imageid'] = code

print(f"✅ Assigned unique 4-digit codes to {len(enriched_df)} movies")
print(f"   Code range: {enriched_df['imageid'].min()} to {enriched_df['imageid'].max()}")

# ============================================
# STEP 3: Image Download Function with Error Handling
# ============================================
def download_poster_with_id(row, save_folder):
    """
    Download poster image for a movie row using 4-digit ID.
    Returns: (success: bool, message: str, saved_path: str or None, file_format: str or None)
    """
    title = str(row['title']).strip() if pd.notna(row['title']) else "unknown"
    image_id = str(row['imageid'])
    poster_url = row.get('Poster', None)

    # Check if Poster field exists and is not null/NaN/empty
    if pd.isna(poster_url) or poster_url is None or poster_url == '' or poster_url == 'N/A':
        return False, "Poster URL is null/empty", None, None

    # Validate URL format
    if not isinstance(poster_url, str) or not poster_url.startswith(('http://', 'https://')):
        return False, f"Invalid URL format: {poster_url}", None, None

    # Get file extension from URL
    parsed_url = urlparse(poster_url)
    path_ext = os.path.splitext(parsed_url.path)[1].lower()

    # Default to .jpg if no extension found or invalid
    if path_ext not in ['.jpg', '.jpeg', '.png', '.webp']:
        path_ext = '.jpg'

    # Remove the dot from extension for format field
    file_format = path_ext[1:]  # e.g., 'jpg' instead of '.jpg'

    # Create filename using 4-digit ID
    filename = f"{image_id}{path_ext}"
    filepath = os.path.join(save_folder, filename)

    try:
        # Download image with timeout
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        }
        response = requests.get(poster_url, headers=headers, timeout=15, stream=True)
        response.raise_for_status()  # Raise error for bad status codes

        # Check if content is actually an image
        content_type = response.headers.get('Content-Type', '')
        if 'image' not in content_type:
            return False, f"Not an image (Content-Type: {content_type})", None, None

        # Save image
        with open(filepath, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)

        # Verify file was saved and is not empty
        if os.path.exists(filepath) and os.path.getsize(filepath) > 0:
            return True, f"Saved: {filename}", filepath, file_format
        else:
            return False, "File saved but empty or corrupted", None, None

    except requests.exceptions.Timeout:
        return False, "Download timeout", None, None
    except requests.exceptions.ConnectionError:
        return False, "Connection error", None, None
    except requests.exceptions.HTTPError as e:
        return False, f"HTTP Error: {e.response.status_code}", None, None
    except requests.exceptions.RequestException as e:
        return False, f"Request failed: {str(e)}", None, None
    except Exception as e:
        return False, f"Unexpected error: {str(e)}", None, None

# ============================================
# STEP 4: Execute Download Pipeline
# ============================================
print("\n" + "="*70)
print(" STARTING POSTER DOWNLOAD PIPELINE WITH 4-DIGIT IDs")
print("="*70)

# Track results
download_results = []
failed_downloads = []
successful_downloads = []
null_posters = []

# Process each row with progress bar
for idx, row in tqdm(enriched_df.iterrows(), total=len(enriched_df), desc="Downloading Posters"):
    success, message, filepath, file_format = download_poster_with_id(row, poster_folder)

    # Update the dataframe with format information
    if success:
        enriched_df.loc[idx, 'format'] = file_format
    else:
        enriched_df.loc[idx, 'format'] = None  # Explicitly set to None if no image

    result = {
        'index': idx,
        'imageid': row['imageid'],
        'title': row['title'],
        'success': success,
        'message': message,
        'filepath': filepath,
        'format': file_format
    }
    download_results.append(result)

    if success:
        successful_downloads.append(result)
    else:
        failed_downloads.append(result)
        # Check if it's a null/empty poster
        if 'null' in message.lower() or 'empty' in message.lower():
            null_posters.append(result)

    # Small delay to be respectful to the server
    time.sleep(0.2)

# ============================================
# STEP 5: Generate Summary Reports
# ============================================
print("\n" + "="*70)
print(" DOWNLOAD SUMMARY")
print("="*70)

total_movies = len(enriched_df)
successful_count = len(successful_downloads)
failed_count = len(failed_downloads)
null_count = len(null_posters)
actual_failures = failed_count - null_count  # Failures that aren't just null posters

print(f"\nTotal Movies Processed: {total_movies}")
print(f"✅ Successfully Downloaded: {successful_count} ({successful_count/total_movies*100:.1f}%)")
print(f"❌ Failed Downloads: {failed_count} ({failed_count/total_movies*100:.1f}%)")
print(f"   - Null/Empty Posters: {null_count}")
print(f"   - Actual Download Failures: {actual_failures}")

# ============================================
# STEP 6: Print Detailed Failure Report with Codes
# ============================================
if failed_downloads:
    print("\n" + "="*70)
    print("❌ POSTERS NOT DOWNLOADED - DETAILED LIST WITH CODES")
    print("="*70)

    # Separate null posters from actual failures
    actual_failure_list = [f for f in failed_downloads if f not in null_posters]

    if null_posters:
        print(f"\n NULL/EMPTY POSTER URLs ({len(null_posters)} movies):")
        print("-" * 70)
        print(f"{'Code':<8} {'Title':<40} {'Status'}")
        print("-" * 70)
        for item in null_posters:
            print(f"{item['imageid']:<8} {item['title'][:38]:<40} {item['message']}")

    if actual_failure_list:
        print(f"\n🚫 DOWNLOAD FAILURES ({len(actual_failure_list)} movies):")
        print("-" * 70)
        print(f"{'Code':<8} {'Title':<40} {'Error'}")
        print("-" * 70)
        for item in actual_failure_list:
            print(f"{item['imageid']:<8} {item['title'][:38]:<40} {item['message']}")
            # Try to show the poster URL if available
            try:
                poster_url = enriched_df.loc[item['index'], 'Poster']
                if pd.notna(poster_url):
                    url_display = str(poster_url)[:50] + "..." if len(str(poster_url)) > 50 else str(poster_url)
                    print(f"{'':<8} {'':<40} URL: {url_display}")
            except:
                pass
        print()

# ============================================
# STEP 7: Print Complete Code-to-Title Mapping
# ============================================
print("\n" + "="*70)
print("📋 COMPLETE IMAGE ID TO TITLE MAPPING")
print("="*70)
print(f"{'Code':<8} {'Title':<45} {'Format':<10} {'Status'}")
print("-" * 70)

for idx, row in enriched_df.iterrows():
    status = "✅ Downloaded" if row['format'] is not None else "❌ No Image"
    format_display = row['format'] if row['format'] is not None else "None"
    title_display = str(row['title'])[:43]
    print(f"{row['imageid']:<8} {title_display:<45} {format_display:<10} {status}")

# ============================================
# STEP 8: Save Download Report to CSV
# ============================================
results_df = pd.DataFrame(download_results)
report_path = os.path.join(poster_folder, "download_report.csv")
results_df.to_csv(report_path, index=False)

# Also save the mapping as a separate CSV for manual updates
mapping_df = enriched_df[['imageid', 'title', 'format']].copy()
mapping_path = os.path.join(poster_folder, "image_id_mapping.csv")
mapping_df.to_csv(mapping_path, index=False)

print("\n" + "="*70)
print("💾 REPORTS SAVED")
print("="*70)
print(f"📄 Full download report: {report_path}")
print(f"📄 Image ID mapping: {mapping_path}")

# ============================================
# STEP 9: Quick Verification
# ============================================
print("\n" + "="*70)
print("🔍 VERIFICATION")
print("="*70)

# Count actual files in folder
saved_files = [f for f in os.listdir(poster_folder) if f.endswith(('.jpg', '.jpeg', '.png', '.webp'))]
print(f"📁 Total image files in '{poster_folder}': {len(saved_files)}")

# Show sample of saved files with their codes
if saved_files:
    print(f"\n Sample of saved files:")
    for f in sorted(saved_files)[:10]:
        file_size = os.path.getsize(os.path.join(poster_folder, f))
        print(f"   • {f} ({file_size:,} bytes)")

# Verify dataframe has the new columns
print(f"\n✅ DataFrame columns updated:")
print(f"   - 'imageid': {enriched_df['imageid'].notna().sum()} codes assigned")
print(f"   - 'format': {enriched_df['format'].notna().sum()} formats recorded, {enriched_df['format'].isna().sum()} None values")

print("\n" + "="*70)
print("✅ POSTER DOWNLOAD PIPELINE WITH 4-DIGIT IDs COMPLETED")
print("="*70)

# ============================================
# STEP 10: Display Final DataFrame Sample
# ============================================
print("\n FINAL DATAFRAME SAMPLE (showing new columns):")
display_cols = ['imageid', 'title', 'format'] + [col for col in enriched_df.columns if col not in ['imageid', 'title', 'format']][:5]
print(enriched_df[display_cols].head(10).to_string())
```
